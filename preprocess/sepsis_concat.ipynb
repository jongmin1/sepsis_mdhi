```python
import pandas as pd
from tqdm.notebook import tqdm
import os
```

```python
filenames = os.listdir('./my_data_1hour/')
filenames
```

```python
def sep_concat(data, filename): 
    name = filename.split('.')[0]
    
    new_data = pd.read_csv('./my_data_1hour/'+filename)
    
    new_data.drop(['Unnamed: 0'], axis = 1, inplace = True)
    new_data.rename(columns = {'valuenum':name}, inplace=True)
    
    if len(data) == 0 : 
        return new_data

    if 'wbc' in filename or 'pH' in filename:
        new_data.drop(['labevent_id'], axis = 1, inplace = True)
        data = pd.merge(data, new_data, how='outer', on=['subject_id', 'hadm_id', 't_time'])
    else :
        data = pd.merge(data, new_data, how='outer', on=['stay_id', 't_time', 'hadm_id', 'subject_id'])
    
    return data
```

```python
data = pd.DataFrame([])

for f in tqdm(filenames):
    if f[0] == '.' : continue;
    data = sep_concat(data, f)

data
```

```python
data.to_csv('all_parameters.csv')
```
